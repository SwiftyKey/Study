```python
import pandas as pd
```

1. Создавйте DataFrames, прочитав файлы https://cloud.ocherepanov.ru/s/gpS7HmfGLpqSSaa и https://cloud.ocherepanov.ru/s/ddgJPBNR2zDJbFL

```python
customers_df = pd.read_csv('customers.csv')
customers_df.head()
```

```python
orders_df = pd.read_csv('orders.csv', decimal=',')
orders_df.head()
```

2. Выведите все заказы клиенты с именем Andrew Allen.

```python
customer_id = customers_df[customers_df['name'] == 'Andrew Allen'].iloc[0]['id']
customer_orders = orders_df[orders_df['customer_id'] == customer_id]
customer_orders
```

3. Соедините таблицы в одну и сохраните в отдельном файле.

```python
merged_customers_orders_df = pd.merge(
    orders_df, customers_df, left_on='customer_id', right_on='id')
merged_customers_orders_df.to_csv('merged_customers_orders.csv', index=False)
merged_customers_orders_df.head()
```

4. Посчитайте количество заказов клиентов из Los Angeles.

```python
len(merged_customers_orders_df[merged_customers_orders_df['city'] == 'Los Angeles'])
```

5. Сгруппируйте заказы по городам и посчитайте их количестово и сумму.

```python
city_stats = merged_customers_orders_df.groupby('city').agg(
    order_count=('sales', 'size'),
    total_sales=('sales', 'sum')
)
city_stats.head()
```

6. Выведите топ-5 людей, котороые больше всего потратили денег.

```python
top_5_customers = merged_customers_orders_df.groupby('customer_id')[
    'sales'].sum().nlargest(5)
top_5_customers
```

7. Посчитайте средний чек по всем штатам за каждый год.

```python
merged_customers_orders_df['order_date'] = pd.to_datetime(
    merged_customers_orders_df['order_date'])
merged_customers_orders_df['year'] = merged_customers_orders_df['order_date'].dt.year
merged_customers_orders_df.head()
```

```python
avg_sales = merged_customers_orders_df.groupby(['state', 'year'])['sales'].mean().reset_index()
avg_sales.rename(columns={'sales': 'avg_sales'}, inplace=True)
avg_sales.head()
```

8. Создайте Dataframe, прочитав файл https://cloud.ocherepanov.ru/s/LArppdaMqACpnpr

```python
winds_df = pd.read_csv('winds.csv')
winds_df = winds_df.sort_values('dt').reset_index(drop=True)
winds_df['group'] = (winds_df['direction'] == 'x').cumsum()
winds_df['dt'] = winds_df.groupby('group')['dt'].transform('min')
winds_df.head()
```

9. С помощью методов unstack и pivot создайте wide-представление, в таком виде: время - высота - скорость по x - скорость по y - скорость по z.

```python
winds_df = winds_df.set_index(['dt', 'direction'])
winds_df.head()
```

```python
wide_pivot_winds_df = winds_df.pivot_table(
    index='dt', columns='direction', values='value')
wide_pivot_winds_df.columns.name = None
wide_pivot_winds_df.rename(
    columns={'x': 'speed_x', 'y': 'speed_y', 'z': 'speed_z'}, inplace=True)
wide_pivot_winds_df.head()
```
