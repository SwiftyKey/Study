7. Создайте объект Series строк. Отберите элементы, которые содержат не менее двух гласных.

```python
s = pd.Series(['apple', 'orongo', 'sky', 'beautiful', 'butterfly'])

vowels = 'aeiouAEIOU'
vowel_count = s.apply(lambda x: str(set(x))).str.count(f'[{vowels}]')
s[vowel_count >= 2]
```

8. Создайте случайный Series. Найти индексы локальных минимумов.

```python
s = pd.Series(np.random.randn(10))
print("Series:\n", s)

local_min = (s < s.shift(1)) & (s < s.shift(-1))
min_indices = s[local_min].index
print("Индексы локальных минимумов:\n", min_indices.tolist())
```

9. Создайте случайный Series. Найдите автокорреляцию.

```python
'''
0: 3 1 4 1 3 2 3 2 4 2
1: n 3 1 4 1 3 2 3 2 4
2: n n 3 1 4 1 3 2 3 2
3: n n n 3 1 4 1 3 2 3
4: n n n n 3 1 4 1 3 2
5: n n n n n 3 1 4 1 3
6: n n n n n n 3 1 4 1
7: n n n n n n n 3 1 4
8: n n n n n n n n 3 1
'''
s = pd.Series(np.random.randint(1, 15, size=100))
print(s)
lags = range(1, 26)
autocorrs = {lag: s.corr(s.shift(lag)) for lag in lags}

print("Автокорреляции:")
for lag, acf in autocorrs.items():
    print(f"   Лаг {lag:2d}: {acf:8.6f}")
```

3.   Переименуйте столбцы так, чтобы они имели русские названия.


```python
df.columns = ['Модель', 'КПП', 'Мощность', 'Цена']
df.head()
```

4.   Выведете только столбец с названием моделей автомобиля.


```python
df['Модель'].head()
```

5.   Получите информацию только о автомобилях марки Nissan.


```python
df['Марка'] = df['Модель'].str.split().str[0]
df[df['Марка'] == 'Nissan']
```

6.   Получите информацию только о автомобилях не старше 2 лет.


```python
df['Год'] = df['Модель'].str.extract(r'(\d{4})').astype(int)
df[df['Год'] >= 2023]
```

7.   Добавьте столбец с ценой автомобилей в рублях.


```python
df = df.rename(columns={'Цена': 'Цена, дол.'})
df['Цена, дол.'] = df['Цена, дол.'].astype(str).str.replace(',', '').astype(float)
df['Цена, руб.'] = df['Цена, дол.'] * 83.35
df
```

8.   Отсортируйте автомобили по стоимости, начиная с дешевых.


```python
df_sorted = df.sort_values('Цена, руб.', ascending=True)
df_sorted
```

9.   Найдите самую дорогую машину.

10.  Отсортируйте автомобили по критерию "Самая дешевая лошадинная сила".


```python
df['Мощность'] = df['Мощность'].str.extract(r'(\d+)').astype(int)
df['Цена за силу'] = df['Цена, дол.'] / df['Мощность']
df_sorted = df.sort_values('Цена за силу', ascending=True)
```
