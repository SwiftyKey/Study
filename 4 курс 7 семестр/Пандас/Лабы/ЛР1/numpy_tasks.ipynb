4. Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы. Для X =
np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])

```python
X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
print(X)
diag = np.diag(X)
print(diag[diag != 0].prod())
```

5. Найти максимальный элемент в векторе x среди элементов, перед которыми стоит нулевой. Для x =
np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])

```python
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
print(x)
prev_is_zero = (x[:-1]==0)
print(prev_is_zero)
print(np.max(x[1:][prev_is_zero]))
```

6. Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами.

```python
X, Y = np.random.randint(0, 10, (2, 5)), np.random.randint(5, 15, (3, 5))

print(X[:, np.newaxis])
print(Y[np.newaxis, :])
print(np.sqrt(((X[:, np.newaxis] - Y[np.newaxis, :]) ** 2).sum(axis=2)))
```
