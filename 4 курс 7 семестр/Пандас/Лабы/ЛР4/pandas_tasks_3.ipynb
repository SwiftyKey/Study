```python
import pandas as pd
```

1. Создать Dataframe на базе csv

```python
df = pd.read_csv('titanic.csv')
df.head()
```

2.   Проверить, существуют ли пропущенные значение в данных.

```python
any(df.isnull())
```

3.   Получить наименование столбцов с пропущенными значениями и их количество

```python
counter = df.isnull().sum()
cols = counter[counter > 0]
cols
```

4.   Заменить пропущенные значение в столбце Age в зависимости от столбца Name. Самостоятельно предложить решение.

```python
df['Title'] = df['Name'].str.extract(r',\s*([^\.]*)\.', expand=False)
median = df.groupby('Title')['Age'].median()
median
```

```python
df['Age'] = df.apply(lambda x: median[x['Title']] if pd.isnull(x['Age']) else x['Age'], axis=1)
df.head()
```

5.   Проверить, присутствуют ли дубликаты в данных по всем столбцам.

```python
df.duplicated().sum()
```

6.   Проверить присутствие дубликатов по разным подмножествам столбцов.

```python
duplicates_name_age = df.duplicated(subset=['Name', 'Age']).sum()
print(f"Дубликаты по Name и Age: {duplicates_name_age}")

duplicates_ticket_pclass = df.duplicated(subset=['Ticket', 'Pclass']).sum()
print(f"Дубликаты по Ticket и Pclass: {duplicates_ticket_pclass}")

duplicates_cabin_sex = df.duplicated(subset=['Cabin', 'Sex']).sum()
print(f"Дубликаты по Cabin и Sex: {duplicates_cabin_sex}")
```

7.   Заменить значение female на 0, а male на 1 в столбце Sex.

```python
df['Sex'] = df['Sex'].map({'female': 0, 'male': 1})
df.head()
```

8.   Выделить в новый признак титулы из имени и посчитать сколько количество Mr и Miss.

```python
ct = df['Title'].value_counts()
print(f'Mr: {ct['Mr']}')
print(f'Miss: {ct['Miss']}')
```

9.   Cоздать новый признак FamilySize (размер семьи) как комбинацию SibSp и Parch

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df.head()
```

10.  Создаем новый признак IsAlone - один или с семьей.  

```python
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
df.head()
```
